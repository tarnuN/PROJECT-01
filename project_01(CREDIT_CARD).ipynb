```python
df.fillna({'V23':'0','V24':'0','V25':'0','V26':'0','V27':'0','V28':'0','Amount':'0','Class':'0'})
```

```python
new_df=df.fillna(method='ffill')
df=new_df
```

```python
#NOR means normal transition
#fraud transition  there unbalanced
nor=df[df.Class==0]
fraud=df[df.Class==1]
```

```python
print(nor.shape)
print(fraud.shape)
```

```python
nor_sam=nor.sample(n=2)
```

```python
new_dataset=pd.concat([nor_sam,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x1=model.predict(x_train)
print(x1)
accuracy = accuracy_score(y_train, x1)
print(accuracy)
```
