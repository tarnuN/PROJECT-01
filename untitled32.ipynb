```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return x**3 - 2*x - 5

def bisection(func, a, b, tol):
    if func(a) * func(b) > 0:
        print("The function does not have opposite signs at interval endpoints.")
        return None

    iterations = []

    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        iterations.append(midpoint)
        if func(midpoint) == 0:
            return midpoint
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint

    return iterations

# User input for function, interval endpoints, and tolerance
func_str = input("Enter your function (use 'x' as the variable): ")
func = lambda x: eval(func_str)
a = float(input("Enter left endpoint of interval: "))
b = float(input("Enter right endpoint of interval: "))
tol = float(input("Enter tolerance: "))

# Perform bisection
root_iterations = bisection(func, a, b, tol)

if root_iterations:
    print("Root:", root_iterations[-1])

    # Plot function
    x_vals = np.linspace(a, b, 100)
    y_vals = func(x_vals)
    plt.plot(x_vals, y_vals, label=func_str)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(root_iterations[-1], color='red', linestyle='--', label='Root')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Bisection Method')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def func(x):
    return x**3 - 2*x - 5

def bisection(func, a, b, tol):
    if func(a) * func(b) > 0:
        print("The function does not have opposite signs at interval endpoints.")
        return None, None

    iterations = 0

    while (b - a) / 2.0 > tol:
        midpoint = (a + b) / 2.0
        iterations += 1
        if func(midpoint) == 0:
            return midpoint, iterations
        elif func(a) * func(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint

    return (a + b) / 2.0, iterations

# User input for function, interval endpoints, and tolerance
func_str = input("Enter your function (use 'x' as the variable): ")
func = lambda x: eval(func_str)
a = float(input("Enter left endpoint of interval: "))
b = float(input("Enter right endpoint of interval: "))
tol = float(input("Enter tolerance: "))

# Perform bisection
root, iterations = bisection(func, a, b, tol)

if root is not None:
    print("Root:", root)
    print("Number of iterations:", iterations)

    # Plot function
    x_vals = np.linspace(a, b, 100)
    y_vals = func(x_vals)
    plt.plot(x_vals, y_vals, label=func_str)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(root, color='red', linestyle='--', label='Root')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Bisection Method')
    plt.legend()
    plt.grid(True)
    plt.show()
```
